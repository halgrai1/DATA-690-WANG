### Step 4: Clean raw_data so that only yearly data files remain


```python
yearly_data = [i for i in raw_data if "MERGED" in i]
yearly_data
```

### Step 5: Create New Dataframe 
* 6 columns: "UNITID", "INSTNM", "STABBR", "REGION", "ADM_RATE", "TUITIONFEE_IN"
* Add Year column #### and convert into integer

```python
# new dataframe is df_list
df_list = []
for file_name in yearly_data:
    df = pd.read_csv(file_name,
                         usecols = ["UNITID", "INSTNM", "STABBR", "REGION", "ADM_RATE", "TUITIONFEE_IN"])
    year = file_name[6:10]
    df["Year"] = year
    df_list.append(df)
```

```python
df_list
```

```python
#concatenating multiple dataframes

df = pd.concat(df_list)
```
