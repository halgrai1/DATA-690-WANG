#### 4. Extract the column "Census Year" from data_list and assign them to a list named year_list. Remove the "" from the last element "2010". Print the cleansed year_list.

```
# splitting list of list using list comprehension
data_split = [i.split("\t") for i in data_list]

## Checkpoint
#print(data_split)
#print(data_list)
#for row in data_split:
 # print(row)
```

```
# creating one list
data_split2 = [x for l in data_split for x in l]

## Checkpoint
print(data_split2)
```

```
# creating a list for census year called year_list
year = data_split2[::4]
year2 = [x.replace("2010*","2010") for x in year]
year_list = [int(y) for y in year2]

print(year_list)

## Checkpoint
#print(type(year_list))
#print(type(year_list[0]))
```

#### 5. Extract the "Total Population" column from the data_list and assign them to a list named "pop_list". Remove the "," from the numbers since Python doesn't recognize them. Print the cleansed "pop_list".


```
# extracting pop column 
pop = data_split2[1::4]

##Checkpoint
print(pop)
```

```
#creating a list for Total Population called pop_list
pop2 = [p.replace(",","") for p in pop]
pop_list = [int(o) for o in pop2]

print(pop_list)
```

####6. Extract the "Census Cost" column from the data_list and assign them to a list named "cost_list". Remove the ",", and "$", and "Billion". Make sure to add the "0"s to the numbers from which you removed "Billion". Print the cleansed cost_list.


```
#extracting census cost column
cen_cost= data_split2[2::4]

## Checkpoint
print(cen_cost)
```

```
# creating a list for Census Cost called cost_list
cen_cost1 = [a.replace(".","") for a in cen_cost]
cen_cost2 = [a.replace(",","") for a in cen_cost1]
cen_cost3 = [b.replace("$","") for b in cen_cost2]
cen_cost4 = [c.replace(" Billion","000000000") for c in cen_cost3]
cost_list = [int(d) for d in cen_cost4]

print(cost_list)
```

####7. Extract the "Average Cost per Person" column from the data_list and assign them to a list named "avg_list". Remove the "cents", and "$". Make sure to divide the numbers in cents by 100 so that all numbers are measured in dollar. Print the cleansed avg_list.


```
# extracting the average cost person column
avg_cost = data_split2[3::4]

# Checkpoint
print(avg_cost)
```

```
# creating a list for the average cost column called avg_list
avg_cost2 = [a.replace("cents\n","") for a in avg_cost]
avg_cost3 = [b.replace(".","") for b in avg_cost2]
avg_cost4 = [c.replace("$","") for c in avg_cost3]
avg_cost5 = [d.replace("\n","") for d in avg_cost4]
avg_cost6 = [int(e) for e in avg_cost5]
avg_list = [f/100  for f in avg_cost6]

#Checkpoint
#print(avg_cost2)
#print(avg_cost3)
#print(avg_cost4)
#print(avg_cost5)
#print(avg_cost6)

#print list
print(avg_list)
```

#### 8. Coalesce the cleansed data and save them to a text file named "census_cost.csv". The new file should look similar to the original source file except that it is in comma-delimited format and the numbers have been cleansed. The top two lines from the original file should be retained in the new file.


```
#creating csv file
import csv

# creating row_list for neater code
row_list = [top2_list, year_list,pop_list,cost_list,avg_list]

with open("census_cost.csv", "w") as f:
  #configure writer to write standard csv file
  writer = csv.writer(f, delimiter=',',lineterminator='\n')
  writer = csv.writer(f)
  writer.writerows(row_list)
```
