```python
weather = pd.read_csv('/content/drive/MyDrive/pydatadc_2018-tidy-master/data/weather.csv')
```

```python
weather
# problem: data is stored in both rows and columns
# when values are duplicated its indicative of this issue
```

```python
# can use pd.melt, or weather.melt
weather_melt = weather.melt(
    id_vars=['id', 'year', 'month', 'element'], #element will be pivoted later
    var_name='day', # we're first fixing the multiple day columns using melt
    value_name='temp'
)
```

```python
weather_melt.head()

# still seeing duplicated rows
```

```python

weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],  # columns we don't want to pivot
    columns='element',   #column to pivot. for each unique value of this value it creates a new column
    values='temp'
)
```

```python
weather_tidy 
# prints as a heirarchical index
```

```python
# flattens dataframe
# sometimes have to run it twice

weather_tidy.reset_index() 
```

```python
billboard_melt.loc[billboard_melt['track'] == 'Loser']
# problem: the song is repeated in this dataset 76 times
# takes up too much storage
```

```python
# creating a dataset that only has information about songs

billboard_songs = billboard_melt[['year', 'artist', 'track', 'time']]
```

```python
billboard_songs.head()
```

```python
billboard_songs.shape
```

```python
billboard_songs = billboard_songs.drop_duplicates()
```

```python
billboard_songs.shape
```

```python
# generate a key called id
# helps with merging data on the column

billboard_songs['id'] = range(len(billboard_songs))
```

```python
billboard_songs.head()
```

```python
# save new data set
billboard_songs.to_csv('billboard_songs.csv', index=False)
```

```python
# merging two dataset
# melt is left; song is right
# use on when column names are the same for both dataset

billboard_ratings = billboard_melt.merge(
    billboard_songs, on=['year', 'artist', 'track', 'time']   #these are the same columns between the dataset
)
```

```python
billboard_ratings.head()

# id is now added one and is unique for each song
```

```python
# creating a final dataset with only the columns we want
# drop duplicates if necessary
# breaking up data into meaningful part

billboard_ratings = billboard_ratings[['id', 'date.entered', 'week', 'rating']]
```

```python
billboard_ratings.to_csv('billboard_ratings.csv', index=False)
```
